```python
import turtle

screen = turtle.Screen()

screen.title("Drawing a Square with Turtle")

# Create the turtle

pen = turtle.Turtle()

# Draw a square

for _ in range(4):

  pen.forward(100) # Move forward 100 units

  pen.right(90)

screen.mainloop()
```

```python
# Assigning Different Variables

name = "Shamonty"

age = 15

is_student = True

weight = 38.5

# Printing Different Variables and their Data Type

print("Name :", name)

print("Data Type of Name is", type(name))

print("Age :", age)

print("Data Type of Age is", type(age))

print("is_student :", is_student)

print("Data Type of is_student is", type(is_student))

print("Weight :", weight)

print("Data Type of weight is", type(weight))

# Type casting to convert the datatype of variables

print("\n After Type Casting....")

age = str(age)

print(age)

print("Data Type of age is", type(age))

weight = int(weight)

print(weight)

print("Data Type of Weight is", type(weight))
```
